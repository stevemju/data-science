# College gratuation data between 2010 and 2012

## Important columns

    - Rank - Rank by median earnings (the dataset is ordered by this column).
    - Major_code - Major code.
    - Major - Major description.
    - Major_category - Category of major.
    - Total - Total number of people with major.
    - Sample_size - Sample size (unweighted) of full-time.
    - Men - Male graduates.
    - Women - Female graduates.
    - ShareWomen - Women as share of total.
    - Employed - Number employed.
    - Median - Median salary of full-time, year-round workers.
    - Low_wage_jobs - Number in low-wage service jobs.
    - Full_time - Number employed 35 hours or more.
    - Part_time - Number employed less than 36 hours.


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv('recent-grads.csv')
recent_grads.head(3)
```

# Various plots : exploring the data
## Scatter plots

```python
recent_grads.plot(x='Sample_size',y='Median',xlim=(-100,4500),ylim=(20000,115000),kind='scatter',title='Median salary and Sample size (number of full time workers)')
plt.show()
```

We can see that students in less popular majors tend to make more money, but they also make the lower salaries. For a sample size of less than 100, median salaries are more disparate and goes from 22000 to 110000 (with a big gap between 75000 and 110000).
Where the sample size is more than 100, salaries tend to be more confined around 40000.

```python
ax = recent_grads.plot(x='Sample_size',y='Unemployment_rate',title='Sample size and Unemployment rate',xlim=(-100,4500),ylim=(-0.01,0.18),kind='scatter')
plt.show()
```

Unemployment rate tend to stabilize (around `0.07` when the sample size increases, just like the salaries in the previous plot. However, for less than 100, it goes from `0.0` to `0.17`.

```python
recent_grads.plot(x='Full_time',y='Median',title='Full time employees and median',xlim=(-15000,260000),ylim=(10000,120000),kind='scatter')
plt.show()
```

As the number of full time employees increases, median salary values seem to stabilize around `38000`.

```python
recent_grads.plot(x='ShareWomen',y='Unemployment_rate',title='Women as share of total and unemployment rate',kind='scatter')
plt.show()
```

There is no apparent link between the share of women and the unemployment rate.

```python
recent_grads.plot(x='Men',y='Median',xlim=(-15000,180000),ylim=(10000,120000),title='Number of male graduates and median salary',kind='scatter')
plt.show()
```

```python
recent_grads.plot(x='Women',y='Median',xlim=(-15000,180000),ylim=(10000,120000),title='Number of female graduates and median salary',kind='scatter')
plt.show()
```

In the two precedent scatter plots, we can see that as the number of female graduates increases, the salaries tend to stabilize around `35000`, whereas for the male graduates, they stabilize around `39000-40000`. Also, for a fixed number of males graduates, there is a bigger gap in salaries than for the same number of female graduates, and the highest salaries are higher for the males graduates.

## Bar plots

## Percentage of women for highest or lowest paying majors

```python
recent_grads['ShareWomenPercent']=100*recent_grads['ShareWomen']

ax1 = recent_grads[:10].plot.bar(x='Major', y='ShareWomenPercent', legend=False)
ax1.set_title('Percentage of women for the 10 HIGHEST paying majors')
ax1.set_ylim(0,100)
ax1.axhline(50,c=(171/255,171/255,171/255),alpha=0.4)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(bottom='off',top='off',left='off',right='off')


ax2 = recent_grads[len(recent_grads)-10:].plot.bar(x='Major', y='ShareWomenPercent', legend=False)
ax2.set_title('Percentage of women for the 10 LOWEST paying majors')
ax2.axhline(50,c=(171/255,171/255,171/255),alpha=0.4)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.tick_params(bottom='off',top='off',left='off',right='off')
plt.show()
```

This is a really interesting plot. It explains a part of the wage gap between man and women : women tend to choose lowest paying college majors.
Let's see if this tendency is keeped if we include not 10, but 25 highest or lowest paying majors.

```python
ax1 = recent_grads[:25].plot.bar(x='Major', y='ShareWomenPercent', legend=False)
ax1.set_title('Percentage of women for the 25 HIGHEST paying majors')
ax1.set_ylim(0,100)
ax1.axhline(50,c=(171/255,171/255,171/255),alpha=0.4)
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_visible(False)
ax1.tick_params(bottom='off',top='off',left='off',right='off')

ax2 = recent_grads[len(recent_grads)-25:].plot.bar(x='Major', y='ShareWomenPercent', legend=False)
ax2.set_title('Percentage of women for the 25 LOWEST paying majors')
ax2.axhline(50,c=(171/255,171/255,171/255),alpha=0.4)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['bottom'].set_visible(False)
ax2.tick_params(bottom='off',top='off',left='off',right='off')
plt.show()
```

This can explain why women earn less is we divide all of the income of all women by the number of women and the same for men, without taking education and other factors into account.
